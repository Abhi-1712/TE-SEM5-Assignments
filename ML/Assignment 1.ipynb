```python
data['Ca'].fillna(value=data['Ca'].mean().round(1), inplace=True) #filling NA values of 'Ca' with mean values
data.isna().sum() #confirming is NA are removed
```

```python
data['Thal'].replace([np.nan], data['Thal'].mode().iloc[0], inplace=True) #filling NA values of 'Thal' with mode values
data.isna().sum() #confirming is NA are removed in 'Thal' column
```

```python
mean_age_of_patient = data['Age'].mean().round(0) #finding mean age of patients and rounding it to integer value
mean_age_of_patient
```
